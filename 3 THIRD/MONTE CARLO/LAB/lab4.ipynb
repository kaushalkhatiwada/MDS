## RANDOM WALK

```python
import numpy as np
import matplotlib.pyplot as plt
import random
def randomwalk1D(n):
    x, y = 0, 0
    timepoints = np.arange(n + 1)
    positions = [y]
    directions = ["UP", "DOWN"]
    for i in range(1, n + 1):
        # Randomly select either UP or DOWN
        step = random.choice(directions)
        
        # Move the object up or down
        if step == "UP":
            y += 1
        elif step == "DOWN":
            y -= 1
        # Keep track of the positions
        positions.append(y)
    return timepoints, positions
```

```python
print(randomwalk1D(10))
```

## Markov Chain Ethrenfest model

```python
import random
import matplotlib.pyplot as plt
```

```python

def ehrenfest_model(num_particles, num_steps):
    # Initialize the number of particles in urn A
    particles_in_A = num_particles
    particles_in_B = 0

    # Lists to store the number of particles in each urn over time
    particles_A = [particles_in_A]
    particles_B = [particles_in_B]

    for _ in range(num_steps):
        # Randomly choose a particle to move
        if random.random() < particles_in_A / num_particles:
            particles_in_A -= 1
            particles_in_B += 1
        else:
            particles_in_A += 1
            particles_in_B -= 1

        # Record the current state
        particles_A.append(particles_in_A)
        particles_B.append(particles_in_B)

    return particles_A, particles_B
```

```python

# Parameters
num_particles = 100
num_steps = 1000

# Run the Ehrenfest model
particles_A, particles_B = ehrenfest_model(num_particles, num_steps)
```

```python

# Plot the results
plt.plot(particles_A, label='BOX A')
plt.plot(particles_B, label='BOX B')
plt.xlabel('Steps')
plt.ylabel('Number of Particles')
plt.legend()
plt.title('Ehrenfest Model Simulation')
plt.show()
```

## Equilibrium Distribution

```python
import numpy as np
```

```python
def stationary_distribution(p, q):
    if p + q == 0:
        raise ValueError("p and q cannot both be zero.")
    
    pi_0 = q / (p + q)
    pi_1 = p / (p + q)
    
    return pi_0, pi_1
```

```python


p = 0.3
q = 0.3
pi = stationary_distribution(p, q)
print(f"Stationary distribution for p={p}, q={q}:  = {pi}")
```
