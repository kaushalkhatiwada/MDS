```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
automobile = fetch_ucirepo(id=10) 
  
# data (as pandas dataframes) 
X = automobile.data.features 
y = automobile.data.targets 

X.head(5)
```

```python
import numpy as np

# replace "?" to NaN
X.replace("Nan", np.nan, inplace = True)
X.head(5)
```

```python
missing_data = X.isnull()
missing_data.head(5)
```

```python
# count missing data in each column

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

```python
missing_data_count = X.isnull().sum()

print(missing_data_count)
```

```python
# Calculate the average of normalized-losses
avg_norm_loss = X["normalized-losses"].astype('float').mean()

print("Average of normalized-losses:", avg_norm_loss)

# Replace missing values in "normalized-losses" column with the mean
X["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)

# Now, X["normalized-losses"] will have missing values replaced with the mean
```

```python
# Calculate the average of bore
avg_bore = X["bore"].astype('float').mean()

print("Average of bore:", avg_bore)

# Replace missing values in "bore" column with the mean
X["bore"].replace(np.nan, avg_bore, inplace=True)

# Now, X["bore"] will have missing values replaced with the mean
```

```python
# Calculate the average of stroke
avg_stroke = X["stroke"].astype('float').mean()

print("Average of stroke:", avg_stroke)

# Replace missing values in "stroke" column with the mean
X["stroke"].replace(np.nan, avg_stroke, inplace=True)

# Now, X["stroke"] will have missing values replaced with the mean
```

```python
#replace the missing 'num-of-doors' values by the most frequent 
X["num-of-doors"].replace(np.nan, 4.0, inplace=True)
```

```python
# simply drop whole row with NaN in "price" column
X.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
X.reset_index(drop=True, inplace=True)
```

```python
# Data standardization for meaningful comparison

# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
X['city-L/100km'] = 235/X["city-mpg"]

# check your transformed data 
X.head()
```

```python
# Data normalization: transforming values of several variables into a similar range
# replace (original value) by (original value)/(maximum value)

X['length'] = X['length']/X['length'].max()
X['width'] = X['width']/X['width'].max()
```
