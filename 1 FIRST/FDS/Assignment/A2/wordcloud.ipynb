Data Scraping to scrape the text content from givrn articles using BeautifulSoup library

```python
import requests
from bs4 import BeautifulSoup
```

```python
import re
```

```python
res = requests.get('https://thehimalayantimes.com/opinion/navigating-nepals-digital-frontier-understanding-cybersecurity-in-the-digital-age-ensuring-data-safety-and-the-role-of-ai')
```

```python
html = BeautifulSoup(res.content, "html.parser")
texts = html.findAll('p')

contents=""
for text in texts:
    contents += text.get_text()
```

```python
contents
```

```python
contents = contents.lower()
```

```python
contents = re.sub(r'\d+','',contents)
```

```python
contents
```

```python
from wordcloud import  WordCloud
```

```python
wc = WordCloud().generate(contents)
```
